### 양방향 연결 리스트 (Doubly linked list) 

* 한 쪽으로만 링크 연결 X , 양쪽으로 연결하도록 하는 것
* 즉, 앞으로 (다음 Node) 또는 뒤로(이전 Node)로 이동 가능 

```python
class Node :
    def __init__(self,item) :
        self.data = item
        self.prev = None    # Node 의 구조 확장, None 초기시 이전 노드(prev)로 이동 가능
        self.next = None
```

* 리스트 처음과 끝에 더미 노드(dummy node) 두기 !
* head 와 tail에 더미 노드를 둠 (연결리스트에선 head에만 더미 노드를 줬었음)

```python
class DoublyLinkedListL:
    def __init__(self,item):
        self.nodeCnt = 0
        self.head = Node(None)   # head 와 tail을 None 으로 둠
        self.tail = Node(None)
        self.head.prev = None
        
        self.head.next = self.tail # head 의 next는 tail로 설정
        self.tail.prev = self.head # tail 의 prev는 head로 설정
        self.tail.next = None
```

```python
# 리스트 순회 코드
def traverse(self):
    result = []
    curr = self.head   # 순회는 head 부터 시작 (그림 : ------->)
    
    # 리스트 순회에서 while 부분만 코드 다름
    # tail도 더미노드가 있기 때문에 curr의 next의 next 조건으로 바꿔야 순방향 코드로 설정 할 수 있음
    # 빈 리스트에서도 적용 가능
    while curr.next.next :
        curr = curr.next
        result.append(curr.data)
        
    return result
```

```python
# 리스트 역순회 코드
def traverse(self):
    result = []
    curr = self.tail    # 역순회는 tail 부터 시작  (그림 : <-------)
    
    # 리스트 역순회도 while 부분만 코드 다름
    # curr의 prev 의 prev 일때로 바꿔줘야함.
    while curr.prev.prev :
        curr = curr.prev
        result.append(curr.data)
        
    return result
```

# L.insertAfter(prev,newNode) - 원소의 삽입
* 1. next = prev.next
* 2. newNode에 prev, next 노드 조정(newNode.prev <- prev, newNode.next <- next)
* 3. prev.next <- newNode
* 4. nodeCnt <- codeCnt + 1

```python
# next = _next

def insertAfter(self,prev,newNode):
    next_ = prev.next
    newNode.prev = prev
    newNode.next = next_
    prev.next = newNode
    next_.prev = newNode
    self.nodeCnt += 1
    return True
```

```python
# 특정 원소 얻어내기 (지난번 것과 똑같은 코드)

def getAt(self,pos) :
    if pos < 0 or pos > self.nodeCnt :
        return None
    
    i = 0
    curr = self.head
    while i < pos :
        curr = curr.next
        i += 1
        
    return crr
```

```python
# 원소의 삽입 (L.insertAt(pos,newNode))


def insertAt(self,pos,newNode):
    if pos < 1 or pos > self.nodeCnt + 1 :
        return False
    
    prev = self.getAt(pos - 1)
    return self.insertAfter(prev,newNode)
```

#### 연습문제 - (10-3) 양방향 연결 리스트 노드 삭제 (popAfter,popBefore,popAt)
